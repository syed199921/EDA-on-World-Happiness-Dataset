```python
import pandas as pd
import glob
import os
```

```python
reports_csv_files = glob.glob("data/raw/data/*.csv") 
print("Reports", reports_csv_files)
```

```python
dfs = []
for file in reports_csv_files:
    year = os.path.basename(file).split('.')[0]
    print('year',year)
    df_year = pd.read_csv(file)
    df_year['year'] = year
    dfs.append(df_year)
df = pd.concat(dfs, ignore_index=True)
```
