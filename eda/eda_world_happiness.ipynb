```python
import pandas as pd
import glob
import os
```

```python
reports_csv_files = glob.glob("data/raw/data/*.csv") 
print("Reports", reports_csv_files)
```

```python
dfs = []
for file in reports_csv_files:
    year = os.path.basename(file).split('.')[0]
    print('year',year)
    df_year = pd.read_csv(file)
    df_year['year'] = year
    dfs.append(df_year)
df = pd.concat(dfs, ignore_index=True)
```

```python
# Canonical mapping dictionary
canonical_map = {
    'Country': 'country',
    'Country or region': 'country',
    'Region': 'region',
    'Happiness Rank': 'happiness_rank',
    'Overall rank': 'happiness_rank',
    'Happiness Score': 'happiness_score',
    'Happiness.Score': 'happiness_score',
    'Score': 'happiness_score',
    'Standard Error': 'standard_error',
    'Economy (GDP per Capita)': 'log_gdp_per_capita',
    'Economy..GDP.per.Capita.': 'log_gdp_per_capita',
    'GDP per capita': 'log_gdp_per_capita',
    'Family': 'family',
    'Social support': 'social_support',
    'Health (Life Expectancy)': 'healthy_life_expectancy',
    'Health..Life.Expectancy.': 'healthy_life_expectancy',
    'Healthy life expectancy': 'healthy_life_expectancy',
    'Freedom': 'freedom',
    'Freedom to make life choices': 'freedom',
    'Trust (Government Corruption)': 'perceptions_of_corruption',
    'Trust..Government.Corruption.': 'perceptions_of_corruption',
    'Perceptions of corruption': 'perceptions_of_corruption',
    'Generosity': 'generosity',
    'Dystopia Residual': 'dystopia_residual',
    'Dystopia.Residual': 'dystopia_residual',
    'Whisker.high': 'whisker_high',
    'Whisker.low': 'whisker_low',
    'Lower Confidence Interval': 'lower_confidence_interval',
    'Upper Confidence Interval': 'upper_confidence_interval',
    'year': 'year'
}
```

```python
def standardize_column_names(df):
    return df.rename(columns=lambda col:canonical_map.get(col, col.strip().lower().replace(" ", "_").replace(".","_")))
```
