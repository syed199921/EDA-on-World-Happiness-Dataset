```python
import pandas as pd

df = pd.DataFrame({
    "Country": ["A", "B", "C"],
    "Year": ["2021", "2022", "2023"],  # should be int
    "GDP": ["12,345", "15,200", "N/A"]  # commas & missing values
})
```

```python
df.info()
```

```python
df
```

```python
df['GDP'] = (df['GDP'].replace({"N/A":None, ",":""}, regex=True).astype(float))
```

```python
df['Year'] = df['Year'].astype(int)
```

```python
df['GDP'].mean()
```

```python
df
```

```python
df['GDP'] = df['GDP'].ffill(axis=0)
```
